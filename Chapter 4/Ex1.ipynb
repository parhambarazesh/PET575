```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Ts=1
Tf=500

qinn=[0.45]#m3/s
h=2
A=4
rho=1000
Kc=0.12
g=9.81
z=1
delta_z=0
harray=np.array([])

qut=[z*Kc*np.sqrt(g*h)]

Qu=np.array([])
Qin=np.array([])

for i in range(1,Tf,Ts):
    if i>=200:
        qinn=[0.5]
    if i>=350:
        qinn=[0.2]
    Qu=np.concatenate((Qu,qut),axis=0)
    Qin=np.concatenate((Qin,qinn),axis=0)
    deltah=(1/A)*(np.subtract(qinn,qut))
    h=h+deltah*Ts
    if h<=0:
        h=0
    if h>2:
        h=2
    qut=z*Kc*np.sqrt(g*h)
    harray=np.concatenate((harray,h),axis=0)

plt.figure(1)
plt.plot(range(1,Tf,Ts),harray)
plt.xlabel('Time [s]');
plt.ylabel('Tank level [m]');
plt.title('Tank level when changing flow in');
plt.grid(color='k')

plt.figure(2)
plt.plot(np.arange(1,len(Qin)+1),Qin,'r')
plt.plot(np.arange(1,len(Qin)+1),Qu,'b')
plt.xlabel('Time [s]');
plt.ylabel('Flow rate');
plt.legend(('Flow rate in','Flow rate out'))
plt.grid(color='k')
```
