```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def lowpassFilter(x,dt,timeConstant):
    
    alpha = dt / (timeConstant + dt)
    
    y=np.zeros(len(x))
    y[0] = x[0]
    n=len(x)

    for i in range(1,n):
        y[i] = alpha * x[i] + (1-alpha) * y[i-1]

    return y

x = np.arange(1,15+0.05,0.05)
y = np.sin(x)

y_noise=np.zeros(281)
for count in range(0,len(x)):
    y_noise[count] = y[count] + 0.3*(np.random.rand()-0.5)

plt.figure(1)
plt.plot(x,y_noise,'r')
plt.plot(x,y,'b')
plt.grid(color='k')
plt.legend(['y=sin(x) with noise','y=sin(x) without noise'])

y_noise_LPF = lowpassFilter(y_noise,1,2)

plt.figure(2)
plt.plot(x,y_noise,'r')
plt.plot(x,y_noise_LPF,'k')
plt.plot(x,y,'b')
plt.legend(['Noisy','LPF$_2$','ture values'])
plt.grid(color='k')

y_noise_LPF1 = lowpassFilter(y_noise,1,2)
y_noise_LPF2 = lowpassFilter(y_noise,1,8)
y_noise_LPF3 = lowpassFilter(y_noise,1,15)

plt.figure(3)
plt.plot(x,y_noise,'r')
plt.plot(x,y_noise_LPF1,'k')
plt.plot(x,y_noise_LPF2,'c')
plt.plot(x,y_noise_LPF3,'g')
plt.plot(x,y,'b')
plt.legend(['Noisy','LPF$_2$','LPF$_8$','LPF$_{15}$','Pure'])
plt.grid(color='k')

plt.figure(4)
plt.plot(x,y_noise,'r')
plt.plot(x,y_noise_LPF1,'k')
plt.plot(x,y_noise_LPF2,'c')
plt.plot(x,y_noise_LPF3,'g')
plt.plot(x,y,'b')
plt.legend(['Noisy','LPF$_2$','LPF$_8$','LPF$_{15}$','Pure'])
plt.axis([0,4,0,1.2])
plt.grid(color='k')
```
