```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

# system parameters
m=1        #mass
k=4        #spring constant
c=0.00001      #damping constant

# initial conditions
x0=1          #initial displacement
v0=-1                   #initial velocity

# external force
w=2.02        #force frequency
F0=1/np.sqrt(2)       #force

#  frequency
wn=np.sqrt(k/m)
xi=c/(2*m*wn)
wd=np.sqrt(1-xi**2)*wn

# amplitudes and phase angle
A2=-F0*c*w/((k-m*w**2)**2+c**2*w**2)
A1=(A2*w**2*m-A2*k)/c/w

#since xi=c/(2*m*wn)<1, it is the underdamping system

X=np.sqrt(x0**2+(v0/wn)**2)
phi=math.atan(v0/(x0*wn))

# simulation
total=100      #total simulation time

x=np.zeros(total)
for i in range(1,total+1):
    x[i-1]=X*np.exp(-xi*wn*i)*np.sin(wd*i+phi)+A1*np.sin(w*i)+A2*np.cos(w*i)

plt.figure(1)
plt.plot(np.arange(1,total+1),x)
plt.grid(color='k')
plt.legend('displacement')
plt.xlabel('time')
plt.ylabel('displacement')
```
