```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# system parameters
m=100      #mass 
k=30       #spring constant

# initial conditions
x0=0       #initial displacement
v0=1       #initial velocity

#  frequency
wn=np.sqrt(k/m)

# simulation
total=100      #total simulation time

x=np.zeros(total)
for i in range(1,total+1):
    x[i-1]=x0*np.cos(wn*i)+v0/wn*np.sin(wn*i)      #displacement of the block

plt.figure(1)
plt.plot(np.arange(1,total+1),x)
plt.grid(color='k')
plt.legend('displacement')
plt.xlabel('time')
plt.ylabel('displacement')

```
