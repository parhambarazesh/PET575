```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

M=np.array([[2,0],[0,1]])
C=np.zeros([2,2])
K=np.array([[9,-3],[-3,3]])
F2=2

omega=3/np.sqrt(2)

U0=F2/((6-omega**2)*(3-2*omega**2))*np.array([6, 2*(9-2*omega**2)])/2

U0=[[1,1]]

dU0=[[0,0]]

X=np.append(U0,dU0,axis=1)

dt=0.01
total=12

for i in range(0,int(total/dt),1):
    dX1=X[i][2]*dt+X[i][0]
    dX2=X[i][3]*dt+X[i][1]
    
    dX3=1/2*(3*X[i][1]-9*X[i][0])*dt+X[i][2]
    dX4=(3*X[i][0]-3*X[i][1]+2*np.cos(omega*i*dt))*dt+X[i][3]
    dX=[[dX1,dX2,dX3,dX4]]
    X=np.append(X,dX,axis=0)
    
plt.figure(1)
plt.plot(np.arange(dt,total,dt),X[2:,0],'r')
plt.plot(np.arange(dt,total,dt),X[2:,1],'b')
plt.legend(['U1','U2'])

plt.figure(2)
plt.plot(np.arange(dt,total,dt),X[2:,2],'r')
plt.plot(np.arange(dt,total,dt),X[2:,3],'b')
plt.legend(['U1$_{dot}$','U2$_{dot}$'])
```
